# ðŸ”® Retail Sales Forecasting with Prophet

This notebook uses Facebook Prophet to forecast daily retail sales based on historical transaction data.

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("mock_kaggle.csv")

# Convert and rename columns for Prophet
df['data'] = pd.to_datetime(df['data'])
df = df.rename(columns={'data': 'ds', 'venda': 'y'})

# Keep only necessary columns
df = df[['ds', 'y']]
df.head()
```

```python
# Initialize and train Prophet model
model = Prophet(daily_seasonality=True)
model.fit(df)
```

```python
# Create future dataframe and forecast
future = model.make_future_dataframe(periods=90)
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Plot forecast
fig1 = model.plot(forecast)
plt.title("Forecasted Retail Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Plot forecast components
fig2 = model.plot_components(forecast)
plt.show()
```

```python
# Export forecast to CSV
forecast.to_csv("sales_forecast.csv", index=False)
print("Forecast exported to sales_forecast.csv")
```
